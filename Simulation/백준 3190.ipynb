```python
from collections import deque

def direc_change(A):
    global dx
    global dy
    global idx
    global direc
    if A == 'L':
        idx += 1
        direc = (dx[(idx)%4],dy[(idx)%4])
        return direc
    elif A == 'D':
        idx -= 1
        direc = (dx[(idx)%4],dy[(idx)%4])
        return direc

n = int(input())
k = int(input())
a = [list(map(int,input().split())) for _ in range(k)]
l = int(input())
b = [list(input().split()) for _ in range(l)]
gezi = [[0]*n for _ in range(n)]
gezi[0][0] = 2
direc = (0,1)
cnt = 0
q = []
t = []
t = deque(t)
q = deque(q)
q.append((0,0))
idx = 1
dx = [1,0,-1,0]
dy = [0,1,0,-1]

for i in range(k):
    gezi[a[i][0]-1][a[i][1]-1] = 3

while True:
    cnt += 1
    for o in b:
        if cnt == int(o[0]) + 1:
            direc_change(o[1])

    tmp = q.popleft()
    x = tmp[0] + direc[0]
    y = tmp[1] + direc[1]
    q.append((x,y))
    if 0<=x<n and 0<=y<n:
        if gezi[x][y] == 3:
            gezi[x][y] = 1
            gezi[tmp[0]][tmp[1]] = 2
            t.append((tmp[0],tmp[1]))
        elif gezi[x][y] == 2:
            break
        else:
            gezi[x][y] = 1
            gezi[tmp[0]][tmp[1]] = 2
            t.append((tmp[0],tmp[1]))
            tmp2 = t.popleft()
            gezi[tmp2[0]][tmp2[1]] = 0
    else:
        break

print(cnt)


```
